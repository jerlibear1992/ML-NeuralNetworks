```python
#knn imputer

from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=3)

X_knn = imputer.fit_transform(X_norm_array)
```

```python
X_knn
```

```python
print(X_knn.shape)
# This shows the shape of our imputed dataset, which its first dimension has 23106 elements (participants)
# And for each participant, there are 922 more elements (which are the values for the 922 variables in our dataset)
```

```python
# Let's now check if there are still any variables missing to see if our imputation worked correctly.
# We should see no more missing values and thus our boolean should return false.

X_knn_sum = np.sum(X_knn)
X_knn_has_nan = np.isnan(X_knn_sum)
X_knn_has_nan
```

```python
# Let's now convert our data first back to pandas data frame so that we may add the column headers back
# We will need to do this to remove our Index variable (0), time variable (1), and outcome variable heart failure (105)
# Doing this is the first step of creating our training dataset X_train

X_knn_arrayconvert = pd.DataFrame(X_knn, columns = np.arange(922))
X_knn_arrayconvert
```

```python
#Now let's remove the columns ID, year, and CHF for preparation of imputation

#First, we will only keep values from baseline (0) and year 3 (0.5).
X_knn_arrayconvert1 = X_knn_arrayconvert[X_knn_arrayconvert[1] != 1.0]
X_knn_arrayconvert1
```

```python
# Now let's drop our 3 columns so that we have only the features used for predictions in our training data.
X_data_initial = X_knn_arrayconvert1.drop(X_knn_arrayconvert1.columns[[0, 1, 105]], axis=1)
```

```python
# We can check that with those 3 columns dropped we have only 919 columns left.

X_data_initial_array = X_data_initial.to_numpy()
X_data_initial_array.shape
```

```python
X_data = X_data_initial_array.reshape(7702, 2, 919)
print(X_data.shape)

# This reshapes our initial training data shape from (15,404, 919)
# to (7702, 2, 919) which is mainly so that the sequential model we will run later can have the appropriate input shape (2, 919).

# To go a bit more in detail here, our initial data was 2 dimensions. First dimension has 15404 elements (aka persons, IDs).
# And in each of those elements/IDs we have 919 more elements (aka values for the 919 features).

# It is then reshaped into a 3-dimensional array, with first dimension of 7702 elements (IDs).
# Each of those 7702 elements has 2 additional elements (2 time points Time 0 and Time 3).
# Within each of those 2 elements/times, there are 919 elements (values for the 919 features).
```

# Now let's create our heart failure outcome dataset (Y)

```python
# Take our deleted columns and now keep only those at year 6 (time = 1.0)
Y = X_knn_arrayconvert[[0, 1, 105]][X_knn_arrayconvert[1] == 1.0]
```

```python
Y
```

```python
# Like our X_train data, we need to convert them also into an numpy array
Y_array = Y[[105]].to_numpy()
display(Y_array)
print(Y_array.shape)
```

```python
# To again properly shape this into an acceptable format so that sequential model can run, we reshape this into a 3D array.
Y_data = Y_array.reshape(7702, 1, 1)
print(Y_data.shape)
display(Y_data)
```

```python
# Let's also see whether we have any more missing values for outcomes.

Y_array_sum = np.sum(Y_data)
Y_array_has_nan = np.isnan(Y_array_sum)
Y_array_has_nan
```
